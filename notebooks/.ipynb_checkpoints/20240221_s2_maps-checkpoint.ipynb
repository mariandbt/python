```python
# %%timeit -r 3

bins = np.linspace(0, 100, 41)
err_distribution = np.zeros(len(bins) - 1)

for i, selected_sens in enumerate(sns_positions.sensor_id):
# for i, selected_sens in enumerate([200]):
    
#     _, _, maps = s2tab.create_response_maps(list_of_ie_file_paths[:1], selected_sens)   
    _, _, maps = s2tab.create_response_maps(list_of_ie_file_paths, selected_sens);   
    
    events, _ = np.histogram(maps[2], bins)    
    
    err_distribution = err_distribution + events
    print(f'{i+1}/{len(sns_positions.sensor_id)}')
```

```python
err_distribution
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(7, 7), constrained_layout=True)
font_size = 22

bin_means = (bins[:-1] + bins[1:])/2

events, bins, bars = ax.hist(bin_means, bins,
                             weights = err_distribution,
                             density=False,
                             label='Mean error [%]',
                             histtype='step')
ax.set_yscale('log')

ax.set_title('Uncertainty in the s2 tables', fontsize = font_size)
ax.set_ylabel('Counts', fontsize = font_size)
ax.set_xlabel('Error in the mean [%]', fontsize = font_size)
# ax.legend()
```
